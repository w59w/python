ФУНКЦИИ


функции - это блок кода, который выполняет определенную задачу и который можно повторно использовать в других частях программы.

формальное оопределение функции:

def имя_функции([параметры]):
    инструкции

```python
def say_hello():
    print("Hello")

say_hello()
```

для вызова:
имя_функции ([параметры])

```python
# локальные функции

def print_massages():
    def say_hello(): print("Hello")
    def say_goodbye(): print("Goodbye")
    say_hello()
    say_goodbye()
print_massages()
```

```python
# организция функций

def main():
    say_hello()
    say_goodbye()

def say_hello():
    print("Hello")

def say_goodbye():
    print("Goodbye")
main()
```

ПАРАМЕТРЫ ФУНКЦИИ

```python
def say_hello(name):
    print("Hello", name)

say_hello("Tom")
```
