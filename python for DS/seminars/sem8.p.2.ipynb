```python
# task 1
list1 = [3, 5, 4, 6, 5, 3, 4, 5, 8, 6, 3]
def even_plus_one(n):
    if n % 2 == 0:
        return n + 1
    else:
        return n
result = []
for num in list1:
    result.append(even_plus_one(num))
print(result)
```

```python
# task 2
list2 = [12, 13, 15, 17, 23, 29, 101, 100, 599, 251]
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) +1):
        if n % i == 0:
            return False
    return True

result = []
for num in list2:
    result.append(is_prime(num))
print(result)
```

```python
# task 5
def median(lst):
    lst = sorted(lst)
    n = len(lst)
    mid = n // 2
    if n % 2 == 1:
        return lst[mid]
    else:
        return(lst[mid - 1] + lst[mid])/2

list_1 = [1, 13, 14, 23, 25, 34, 24, 14, 100, 100000]
list_2 = [3, 5, 4, 45, 4, 2, 345, 456, 43456, 33, 1]

print(median(list_1))
print(median(list_2))
```

```python
# task 6
def count_d(n):
    if n <= 1:
        return 0
    count = 0
    for i in range(2, n):
        if n % i == 0:
            count += 1
    return count

def find_c(a,b):
    if a <1:
        a = 1
    result = []
    for n in range(a, b + 1):
        if n <= 1:
            continue
        num_prop = count_d(n)
        if num_prop == 2:
            result.append(n)
    return result
test_range_a = 174457
test_range_b = 174505

test_1 = 1
test_b = 30
special_num = find_c(test_1, test_b)
print(f"primer ot 1 do 30: {special_num}")
```

```python
# task 7
def get_proper_d(n):
    if n <= 1:
        return []
    divisors = []
    for i in range(2,n):
        if n % i == 0:
            divisors.append(i)
    return divisors
def find_n(a,b):
    if a < 1:
        a = 1
    result_dict = {}
    for n in range(a, b + 1):
        if n <= 1:
            continue
        proper_divs = get_proper_d(n)
        if len(proper_divs) == 4:
            result_dict[n] = proper_divs
    return result_dict
test_a = 185311
test_b = 185330
sp_num_7 = find_n(test_a, test_b)
print(f"resultat ot {test_a} do {test_b}: {sp_num_7}")
```

```python
# task 8
def laptop_assistant():

    brand_str = 'Acer, Asus, Honor, HP, Lenovo'
    price_str = '20 000, 35 000, 50 000, 40 000, 25 000'

    I5_mult = 1.20
    i7_mult = 1.45

    brands = [b.strip() for b in brand_str.split(',')]
    prices = [int(p.strip().replace(" ", "")) for p in price_str.split(",")]

    base_prices = dict(zip(brands, prices))
    user_brand = input("Введите название бренда ноутбука (Acer, Asus, Honor, HP, Lenovo): ").strip()

    if user_brand not in base_prices:
        print(f"\n ❌ Извините, бренд '{user_brand}' отсутствует в нашем ассортименте.")

    base_price = base_prices[user_brand]
    price_i5 = round(base_price * I5_mult)
    price_i7 = round(base_price * i7_mult)

    f_i3 = f"{base_price:,}".replace(",", " ")
    f_i5 = f"{price_i5:,}".replace(",", " ")
    f_i7 = f"{price_i7:,}".replace(",", " ")

    print(f"\n ✅ Расчет стоимости для бренда {user_brand}:")
    print(f"{user_brand}")
    print(f"i3: {f_i3} som")
    print(f"i5: {f_i5} som")
    print(f"i7: {f_i7} som")
```

```python
# task 9
import numpy as np
import math
H = 1e-6
def derivative_decorator(func):
    def wrapper(x: float) -> float:
        f_plus_h = func(x + H)
        f_minus_h = func(x - H)
        derivative = (f_plus_h - f_minus_h) / (2 * H)
        return derivative
    return wrapper
def sin_(x: float) -> float:
    def log_(x: float) -> float:
        return np.log(x)
@derivative_decorator
def derivative_sin(x: float) -> float:
    return np.sin(x)
@derivative_decorator
def derivative_log(x: float) -> float:
    return np.log(x)

test_x = math.pi / 4
print(f"точка тестирования x = {test_x:.4f} ---")
cos_x = np.cos(test_x)
numerical_d_sin = derivative_sin(test_x)
print(f"1. y = sin(x)")
print(f"аналитическая (cos(x)): {cos_x:.6f}")
print(f"численная (декоратор): {numerical_d_sin:.6f}")
print(f"разница: {abs(cos_x - numerical_d_sin):.2e}\n")
print(f" разница: {abs(cos_x - numerical_d_sin):.2e}\n")

if test_x > 0:
    one_over_x = 1 / test_x
    numerical_d_log = derivative_log(test_x)

    print(f"2. y = log(x) (ln(x))")
    print(f"   Аналитическая (1/x): {one_over_x:.6f}")
    print(f"   Численная (декоратор): {numerical_d_log:.6f}")
    print(f"   Разница: {abs(one_over_x - numerical_d_log):.2e}")

else:
    print("2. y = log(x): Тестирование невозможно, так как ln(x) не определен для x <= 0.")
```

```python
# task 10
import numpy as np

def gradient(func):
    def wrapper(x, y, h=1e-5):
        dz_dx = (func(x + h, y) - func(x, y)) / h
        dz_dy = (func(x, y + h) - func(x, y)) / h
        return dz_dx, dz_dy
    return wrapper

@gradient
def f(x: float, y: float) -> float:
    return (
        -5 * np.exp(x) / (1 + np.exp(x))**2
        -10 * np.exp(y) / (1 + np.exp(y))**2
    )
x, y = 1.0, 2.0
grad = f(x,y)

print("∂z/∂x =", grad[0])
print("∂z/∂y =", grad[1])
```
