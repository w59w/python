```python
# task 22
"""Нам дан список с самыми популярными брендами. Программа запрашивает у пользователя ввести какой-либо бренд. Если этот бренд встречается в списке, то программа пишет сообщение:

"Данный бренд уже есть в списке"
Если же такого бренда нет, то программа добавляет его в список и затем выводит данный список на отображение.

Примечание:

Пользователь может вводить название и с маленькой буквы"""
brand = ['Acer', 'HP', 'Lenovo', 'Honor']
input_ = input("Vvedite nazvanie brenda: ").capitalize()

if input_ in brand:
    print("takoi brand uje est v spiske")
else:
    brand.append(input_)
    print(f"dobavili novyi brand: {brand}" )
```

```python
# task 23
"""Пользователь вводит два числа через пробел, которые являются границами множества. Составте список всех чисел в включительных границах данного множества. Если второе число меньше первого, то множество будет убывающим.

Если пользователь ввел через пробел одинаковые числа, то список будет состоять из единственного такого числа."""

input_ = int(input("vvedite dva chisla cherez probel: "))
split1 = input_.split()
list23 = []
if split1[0] > split1[1]:
    list23.append(range(split1[0]), split1[1])
elif split1[0] == split1[1]:
    print(list23.append(split1[0]))
else:
    list23.append(range(split1[1]), split1[0])
```
