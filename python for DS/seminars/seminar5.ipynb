```python
# task 2
"""Пользователь вводит с клавиатуры слово и число через пробел. Напишите программу, которая распечатывает данное слово в количестве, равное введенному после пробела числу."""
user_input = input("enter a word and a digit: ")
user_input = user_input.split()
word = user_input[0]
num = int(user_input[1])
for i in range(num):
    print(word, end = " ")
```
