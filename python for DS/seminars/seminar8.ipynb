```python
# task 11
import math
def roots(a, b, c):
    D = b**2 - 4*a*c

    if D < 0:
        return []
    elif D == 0:
        x = -b / (2*a)
        return [x]
    else:
        x1 = (-b + math.sqrt(D))/(2*a)
        x2 = (-b - math.sqrt(D))/(2*a)
        return [[x1, x2]]
print(roots(2, 4, 6))
```

```python
# task 12
def iter_roots(a, b, c):
    roots = []

    for x in range(-1000, 1001):
        if a*x*x + b*x + c == 0:
            roots.append(x)
    return roots
print(iter_roots(1, -3, 2))
```

```python
# task 13
def iter_roots(a,b,c,step):
    roots = []
    x = -100
    eps = 1e-6
    while x<= 100:
        value = a*x*x + b*x + c
        if abs(value) < eps:
            roots.append(round(x,5))
        x+=step
    return roots
```

```python
# task 14
def correct(phone: str)->str | None:
    digits = ''.join(ch for ch in phone if ch.isdigit())
    if digits.startswith('996') and len(digits) == 12:
        digits = digits[3:]
    elif digits.startswith('0') and len(digits) == 10:
        digits = digits[1:]

    if len(digits) != 9:
        return None

    code = digits[:3]
    num1 = digits[3:6]
    num2 = digits[6:]
    return f"+996({code}){num1}-{num2}"
clients = [
    '0777784542',
    '996(707) 435493',
    '0(999)34-54-45',
    '0777-349-034',
    '(707)_ 989989',
    '0453345034',
    '+996777784541'
]

for c in clients:
    print(c, '→', correct(c))
```

```python
# task 16
def str_atrib(s: str):
    res = {
        "count": len(s),
        "upper": 0,
        "lower": 0,
        "digit": 0
    }

    for ch in s:
        if ch.isupper():
            res["upper"] += 1
        elif ch.islower():
            res["lower"] += 1
        elif ch.isdigit():
            res["digit"] += 1
    return res

print(str_atrib('AbaaBBaB123'))
```

```python
# task 17
def visitors_stat(users: dict):
    min_visits = min(users.values())
    max_visits = max(users.values())

    rare = None
    frequent = None

    for user_id, visits in users.items():
        if visits == min_visits:
            rare = user_id
        if visits == max_visits:
            frequent = user_id
    return {"rare": rare, "frequent": frequent}

users = { 'ID_1234': 12,
          'ID_25343': 13,
          'ID_4564': 34,
          'ID_6644': 23,
          'ID_3453': 45,
          'ID_45645': 23,
          'ID_5656': 70,
          'ID_43456': 34,
          'ID_57567': 45,
          'ID_6745': 57,
          'ID_6756': 36,
          'ID_45654': 34}

print(visitors_stat(users))
```

```python
# task 18
def reformat_id_rating(data: dict):
    ids = data["id"]
    ratings = data['rating']

    new_dict = {}
    for i in range(len(ids)):
        new_dict[ids[i]] = ratings[i]

    return new_dict

id_rat_dict = {
    "id": ['ID_123', 'ID_234', 'ID_235', 'ID_234', 'ID_345', 'ID_456',
           'ID_67', 'ID_567', 'ID_4563', 'ID_560'],
    "rating": [3.1, 4.6, 5, 4.8, 3.4, 3.1, 2.9, 1.1, 2.7, 4]
}

print(reformat_id_rating(id_rat_dict))
```

```python
import math

def triangle_attr(a, b):

    gip = math.sqrt(a**2 + b**2)


    small = min(a, b)
    large = max(a, b)


    min_angle_rad = math.atan(small / large)


    min_angle = math.degrees(min_angle_rad)
    max_angle = 90 - min_angle

    return {
        'gip': gip,
        'min_angle': min_angle,
        'max_angle': max_angle
    }
print(triangle_attr(3, 4))
```
