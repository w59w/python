```python
# task 11
import math
def roots(a, b, c):
    D = b**2 - 4*a*c

    if D < 0:
        return []
    elif D == 0:
        x = -b / (2*a)
        return [x]
    else:
        x1 = (-b + math.sqrt(D))/(2*a)
        x2 = (-b - math.sqrt(D))/(2*a)
        return [[x1, x2]]
print(roots(5, 6, 3))
```
