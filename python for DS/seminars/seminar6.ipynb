```python
# task 13
from math import sqrt, asin, degrees, pi
katet_1 = [3, 5, 4, 6, 7, 9, 3, 4, 1, 5]
katet_2 = [6, 4, 5, 2, 7, 9, 1, 5, 4, 3]

hyp_list = []
tan_min_angle = []
sin_min_angle = []
angle_min_deg = []
area_tri = []
area_circ = []

for a, b in zip(katet_1, katet_2):
    #  gipotenuza
    c = sqrt(a**2 + b**2)
    hyp_list.append(c)

    # menshyi i bolshyi katet
    small = min(a, b)
    big = max(a, b)

    #  2. tangens men'shego  ugla
    tan_val = small / big
    tan_min_angle.append(tan_val)

    #  3. sin menshego ugla
    sin_val = small / c
    sin_min_angle.append(sin_val)

    #  4. sam ugol v gradusah
    angle = degrees(asin(sin_val))
    angle_min_deg.append(angle)

    #  5. ploshad treugolnika
    s_tri = a * b / 2
    area_tri.append(s_tri)

    #  6. ploshad opisannoi okrujnosti
    R = c / 2
    S_circ = pi * R**2
    area_circ.append(S_circ)

print(hyp_list)
print(tan_min_angle)
print(sin_min_angle)
print(angle_min_deg)
print(area_tri)
print(area_circ)
```

```python
# task 14

cw_1 = [2, 4, 4, 5, 3, 3, 5, 4, 2, 3]
cw_2 = [5, 4, 4, 3, 4, 3, 4, 5, 3, 5]
cw_3 = [3, 4, 4, 4, 5, 5, 4, 3, 3, 2]

def quarter_grade(marks):
    avg = sum(marks) / len(marks)
    base = int(avg)
    if avg - base >= 0.7:
        return base + 1
    return base

print("CW1:", quarter_grade(cw_1))
print("CW2:", quarter_grade(cw_2))
print("CW3:", quarter_grade(cw_3))
```

```python
# task 17
list_ = [1, 3, 5, 4, 2, 6, 'a', 'b', 'c', 'd', 'q']
even = [list_[i] for i in range(len(list_)) if i % 2 == 0 and isinstance(list_[i], (int, float))]
odd = [list_[i] for i in range(len(list_)) if i % 2 != 0 and isinstance(list_[i], str)]
even_sorted = sorted(even)
odd_sorted = sorted(odd)

res = []
ei = 0
oi = 0

for i in range(len(list_)):
    if i % 2 == 0:
        res.append(even_sorted[ei])
    else:
        res.append(odd_sorted[oi])
print(res)
```

```python
# task 19
list_ = [150, 149, 178, 170, 200, 198, 189, 161, 171, 164, 195, 186]
x_mean = sum(list_) / len(list_)
variance = sum((x - x_mean) ** 2 for x in list_) / len(list_)
sigma = variance ** 0.5
standartized = [(x - x_mean) / sigma for x in list_]
print(standartized)
```
