```python
import pandas as pd
import numpy as np
file_path = r"C:\Users\Пользователь\PycharmProjects\python\python for DS\lessons\bank_accounts_big.xlsx"  # сохраняем путь к файлу
df_acc = pd.read_excel(file_path, sheet_name="accounts")  # читаем лист аккаунтс
df_tx = pd.read_excel(file_path, sheet_name="transactions")  # читаем лист трансакшнс
df_acc.head()
```

```python
df_acc['open_date'] = pd.to_datetime(df_acc['open_date'])
df_tx['tx_date'] = pd.to_datetime(df_tx['tx_date'])
df_acc['balance_kgs'] = pd.to_numeric(df_acc['balance_kgs'], errors='coerce')
df_tx['amount_kgs'] = pd.to_numeric(df_tx['amount_kgs'], errors='coerce')
```

```python
df_active = df_acc[df_acc["status"] == "Active"]
df_active.head(10)
```

```python
df_acc["vip_flag"] = df_acc["balance_kgs"] >= 200000  # новый столбец показывается вип клиент или нет
df_acc["tenure_days"] = (pd.Timestamp("2026-01-08") - df_acc["open_date"]).dt.days  #  количесвто дней пользования
df_acc["income_per_tx"] = (df_acc["income_kgs_last30d"] / df_acc["tx_count_last30d"].replace(0, np.nan)).round(2)  # столбец обещго дохода за 30 дней
df_acc[["client_id", "vip_flag", "tenure_days", "income_per_tx"]]  # проверка создались ил новые столбцы
```

```python
city_status = df_acc.groupby("city").agg(
    clients=("client_id", "nunique"),
    total_balance = ("balance_kgs", "sum"),
    avg_balance = ("balance_kgs", "mean"),
    vip_count = ("vip_flag", "sum"),
).reset_index()
city_status.sort_values("total_balance", ascending=False).head(10)
```

```python
pivot_seq_city = pd.pivot_table(  # создаем сводную таблицу
    df_acc,  # источник данных
    values = "balance_kgs",  #  что агрегирует
    index = "city",  # строки
    columns = "segment",  # колонки
    aggfunc = "mean",  # среднне
).round(2)  # округляем
pivot_seq_city.head()  # смотрим первые строки
```

```python
df_tx_pos = df_tx[df_tx["tx_type"] == "POS"]
df_tX_refund = df_tx[df_tx["tx_type"] == "REFUND"]
tx_sum_by_type = df_tx.groupby("tx_type")["amount_kgs"].sum().sort_values(ascending=False)
tx_sum_by_type = (
    df_tx
    .groupby("tx_type")["amount_kgs"]
    .sum()
    .round()
    .astype("int64")
)
tx_sum_by_type
```
