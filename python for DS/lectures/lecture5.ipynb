Циклы. Многие задачи, которые реализованы в коде имеют в своей основе идею Итерации - когда программа, или алгоритм перебирает значения из какого-либо множества и выполняет набор каких-либо повторяющихся операций.

циклы могут быть ограничены оъемом какого-либо логического условия. но циклы могут быть бесконечными, как в указанном ниже примере:

```python
while True:
    print("Да когда же это закончится...?")
```

With loop
for i in {итерируемое множество}:
    f(i) - некоторое действие
    
где:
i - переменная итерации

in - ключевое слово, указывающее на множество, по которому мы итерируемся, то есть элементы которого мы перебираем.

{...} - само множество. в роли данного множества может быть все что угодно: строка. список. словарь. числовое множество range() и тд.

Ниже представлены примеры, когда цикл ограничесн содержимым какого-либо множества. В роли такого множества может выступать:
- числовое множество range()
- строка (множество символов)
- список (list)
- словарь (dict)
- другие типы множеств

```python
for i in range(1, 5):
    print(i, end=" ")
```

```python
x = "abcdef"
for symbol in x:
    print("str: " + symbol)
```

```python
y = ["a", "b", 'c', 'd']
for i in y:
    print(i)
    print(type(i))
```

```python
for _ in range(7):
    print("Plov")
```

```python
x1 = [1, 2, 3, 4, 5]
for ind in range(len(x1)):
    x1[ind] *= 2.5
```

```python
# task 1
"""Дан список слов. Посчитаем, сколько в списке слов состоящих из трех букв."""
words = ['elephant', 'fly', 'computer', 'planet', 'guitar', 'dog',
         'box', 'car', 'keyboard', 'sky', 'ant', 'ocean', 'street',
         'cat', 'window', 'river', 'forest', 'sun', 'giraffe', 'mountain']
n = 0  # счетчик
for word in words:
    if len(word) == 3:
        n = n + 1
        
print(f"В списке {n} слов состоящих из трезх букв!") 
```

```python
# taask 3
"""Рассмотрим задачу, где нам необходимо вычислить сумму чисел из двух разных листов.
Вначале проитерируемся по индексам. Так как списки равной длины, то у них общая индексация. Можем просто перебрать индексы, и по соответствующему индексу достать соответствующий элемент из первого и второго списка."""
list_1 = [1, 2, 3, 4, 5]
list_2 = [2, 3, 4, 5, 6]
sum_list = []
for i in range(len(list_1)):
    summa = list_1[i] + list_2[i]
    print(list_1[i], "+", list_2[i], "=", summa)
    sum_list/append(summa)
    
"""Второй способ. итерация по ииндексам не всегда является хорошим способом в виду неоптимальных ресурсозатрат. Программа будет чаще всего работать быстрее, елси итерироваться по самим объектам списка или списков. Переберем пары элементов непосредственно, итерируясь по двум спискам одновременно. В этом случае у нас будет 2 переменных итерации. Функция zip() дает нам возможность задать несколько итерируемых объектов. """

for i, j, k in zip(list_1, list_2, list_2):
    print(i, j, k)
    
# сложим все суммы в новый список "sum_list"
sum_list = []
for i , j in zip(list_1, list_2):
    sum_list.append(i + j)
    
print(sum_list)
```

```python
# task 4
# соединим имена и фаимлии

names = ['Alexander', 'Napoleon', 'Lao']
surnames = ['Makedonskiy', 'Bonapart', 'Tzu']
fio = []
for i in range(len(names)):
    x = names[i] + " " + surnames[i]
    fio.append(x)
print(fio)

# drugoi variant
fio = []
for n, s in zip(names, surnames):
    fio.append(n + " " + s)
print(fio)
```

```python
# task 5
"""Найдем минимум функции y = x**2 - 6x + 4
Переберем значения х на достаточно большом множестве и для каждого вычислим значение функции. Сложим все значения в список и выберем наименьшее."""

values = []
for x in range(-100, 100):
    y_temp = x**2 - 6*x + 4
    values.append(y_temp)
    
print(values[:10])
print(min(values))
```
