Dict()
словарь -- изменяемый типа неупорядоченных данных, который состоит из пар элементов.
KEY : VALUE # ключ:значение
Отличительной особенностью является наличие фигурных скобок и фигурных скобок и парных элемнтов через двоеточие.
{
    key_1 : value_1,
    key_2 : value_2,
    key_3 : value_3,
    ....
}
Словари очень частоо используются при решении огромного спектра задач как в работе с данными так и во многих других областях IT. Например сырые данные, которые мы можем получать при API запросе на какой-либо сервер часто имеют расширение JSON, который есть ни что иное как сложная вложенная структура массивов и словарей.

1. Конструкторы и базовые методы:
- __new__ и __init__ - методы для создания нового словаря
- __вщс__ - документацияб описывающая как создавать словари разными способами
2. Основные операции словарем:
- __getitem__ - получение значения по ключу (операция dict[key])
- __setitem__ - установка значения по ключу (операция dict[key])
- __delitem__ - удаление элемента (оперция del dict[key])
- __len__ - получение длины словаря(операция len(dict))
3. Полезные методы для работы:
- get() - безопасное получение значения с возможностью указать значение по умолчанию
- keys() - получение всех ключей
- values() - получение всех значений
- items() - получение пар ключ-значение
- update() - обновление словаря данными из другого словаря
- pop() - удаление и возрат значения по ключу
- clear() - очистка словаря
- copy() - создание копии словаря
4. Операторы сравнения:
__eq__, __ne__, __lt__, __le__, __gt__, __ge__ - для сравнения словарей

```python
print(my_dict_2)
print(my_dict_2.keys())
print(my_dict_2.values())
for i in my_dict_2.values():  # iteration by values
    print(i)

for key in my_dict_2:
    print(key)

# iteration by keys and values
for k, v in zip(my_dict_2.keys(), my_dict_2.values()):
    print(k, "-", v)
```

```python
a = [1213, 2342, 2454, 6535, 5674, 8567, 9876]
b = ['Vasya', 'Masha', 'Kate', 'Iris', 'Petr', 'Denis', 'Karina']
dict_4 = {}
for key, val in zip(a, b):
    dict_4[key] = val

dict_5 = {}
for i, j in enumerate(b):
    dict_5[j] = i
```

enumerate() позволяет проитерироваться по множеству и предоставить две переменные итерации - индекс элемента и сам элемент.

dict comprehension
{key: value for key, value in
f({множество}) if {условие}}


```python
a = [1213, 2342, 2454, 6535, 5674, 8567, 9876]
b = ['Vasya', 'Masha', 'Kate', 'Iris', 'Petr', 'Denis', 'Karina']
```

```python
dict_6 = {f'ID_{id}': name for id, name in zip(a,b)}
dict_6

dict_7 = {name:f'ID_{id}' for id, name in zip(a,b) if name != "denis"}
dict_7
```

```python
c = [[1213, 'Vasya'], [2342, 'Masha'], [2454, 'Kate'], [5674, 'Petr']]

dict_8 = {id: name for [id, name] in c}
dict_8

"Kate" in dict_8
"Vasya" in dict_8.values()
9876 in dict_8
```
