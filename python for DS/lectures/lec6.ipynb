```python
# 1
"""даны номера id пользователей. приведем их к виду 'id_00'"""
id_ = [23, 24, 35, 36, 37, 58, 78, 79, 70]
user_id = ['ID_' + str(i) for i in id_]
print(user_id)
```

```python
# 2
"""Даны номера ID пользователей. Приведем их к виду 'ID_0000'"""
id_ = [1, 4, 23, 24, 35, 36, 37, 101, 293, 1023, 73]
user_id = ['ID_' + ("0000" + str(i))[-4:] for i in id_]
print(user_id)
```

```python
# 3
"""Найдем целые корни квадратного уравнения итеративном методом."""
# x ** 2 - 12 * x + 32 = 0
# classical method
roots = []
for x in range(-100, 100):
    if x**2 - 12*x + 32 == 0:
        roots.append(x)
print(roots)

# list comprehensiom
roots = [x for i in range(-100, 100) if x**2 - 12*x + 32 == 0]
print(roots)
```

```python
# 4
"""Даны два числовых списка с замерами температур воздуха. Первый список - утром, второй - вечером. Для каждой пары отберите минимальное значение температуры и сложите их в новый список."""

morning = [12, 15, 21, 24, 17, 19, 20, 10]
evening = [15, 20, 22, 15, 16, 20, 13, 5]

dit = [min(m, e) for m, e in zip(morning, evening)]
print(dit)
```
