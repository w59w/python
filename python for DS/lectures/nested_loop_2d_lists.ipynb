```python
n = 16
res = []
for i in range(1, n, int(n**.5)):
    temp=[]
    for j in range(i, i+int(n**.5)):
        temp.append(j)
    res.append(temp)
print(*res, sep='\n')
```

```python
def matrix(n):
    return [[j for j in range(i, i + int(n**0.5))]
            for i in range(1, n, int(n**0.5))]
print(matrix(4))
print(matrix(16))
```

```python
"""
найти сумму каждого ряда отдельно
найти сумму кайждой колонны отдельно
найти сумму 2 диагоналей"""

rama = [
    [22, 12, 18, 87],
    [88,17,9,25],
    [10,24,89,16],
    [19,86,23,11]
]
row_sums = []
for row in rama:
    row_sums.append(sum(row))
print("summy ryadov: ", row_sums)
col_sums = []
for col in range(len(rama[0])):
    total = 0
    for row in rama:
        total += row[col]
    col_sums.append(total)
print("summy colon: ", col_sums)
main_diag = 0
for i in range(len(rama)):
    main_diag += rama[i][i]
side_diag = 0
n = len(rama)
for i in range(n):
    side_diag += rama[i][n-1-i]
print("side diagonal: ", side_diag)
print("main diagonal: ", main_diag)
```
