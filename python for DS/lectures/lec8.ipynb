function with if

решим более сложную задачу
напишем функцию, которая не получает на вход ни одного аргумента, но запрашивает числа с клавиатуры через пробел. функция должна сформировать список только четных чисел.

```python
def get_even():
    str_num = input()
    temp_list = str_num.split()
    res_list = []
    for i in temp_list:
        if int(i) % 2 == 0:
            res_list.append(int(i))
    return res_list
get_even()
```

```python
def get_even()-> list:
    return [int(i) for i in input().split() if int(i)%2 == 0]
get_even()
```

return with for

local and global
 в питоне существует два основынх типа переменных: локальные и глабльные

локальные переменные объявляются и используются только внутри определнной функции или блока кода. они видимы только внутри этой области видимости (scope).

локальные переменные имеют более высший приоритет, чем глобальные переменные с тем же именем. если внутри функции объявлена локальная переменная с тем же именем. если внутри функции объявлена локальная переменная с определенным имененм то при обращении к этому имени внутри функции будет использоваться локальная переменная, а не глобальная. локальнаые переменные создаются, когда им присваивается сначение внутри функции и уничтожаются при завершении функции.

```python
x = "akylai"
def my_function():
    x = 10
    print(x)

my_function()
print(x)
```

глобальные переменные:
- глобальные переменные объявляются вне функций и блоков кода и могут быть использованы в любой части программы.
- глобальные переменные видимы во всей программе но их изменение внутри функции требует использования ключевого слова глобал.
- глобальные переменные создаются при присваивании значения за пределами функций и блоков кода и остаются в пмяти на протяжении всей программы.

```python
x = 10
def my_function():
    global x
    x = 20
    print(x)
my_function()
print(x)
```

примечание для цикла for
елси переменная создается внутри цикла (например, for или while), она обычно считается локальной переменной, принадлежащей блоку коода цикла и видимой только внутри этого бблока кода цикла. это означает, что эта переменная не будет доступна за передлами цикла, и ее область видимости ограчинивается этим циклом.

```python
for i in range(5):
    t = i
print(t)
```

итератор map
map(function. object)
map стоит применять к итерируемым объектам, таким как список

```python
def f(x):
    x = x.upper()
    return(x)
a = ['a', 'd', 'g', 'r', 's']
new = []
for i in a:
    new.append(f(i))
list(map(f, a))

def f_2(x):
    if x%2 != 0:
        return x+1
    else:
        return x

list_ = [1, 2, 3, 4, 5, 6, 7]
new = []
for i in list_:
    new.append(f_2(i))

list(map(f_2, list_))
```

```python
def counter(cur=0):
    def step():
        nonlocal cur
        cur += 1
        return cur

    def reset():
        nonlocal cur
        cur = 0
        return cur
    return step, reset()
counter_1 = counter()
counter_2 = counter(2)

counter_1[0](), counter_2[0]()
```

декораторы функций
объявим декоратор. его смысл в том что он является функцией которая дополняет функционал какой либо сторонней функции.

```python
def func_decorator(func):
    def wrapper():
        "я учу кота делать интересные вещи!"
        print("---- *кот смотрит на тебя ---")
        func()
        print("--- *кот делает шкодливый вид и убегает ---")
    return wrapper

def some_cat():
    print("--- Привет! меня зоовут барсик и я тебе сделал кусь! ---")

print(some_cat())
print(func_decorator(some_cat))

@func_decorator
def some_cat():
    print("--- Привет! Меня зовут барсик и я тебе сделал кусь! ---")
some_cat()
```
