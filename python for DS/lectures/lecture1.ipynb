lecture 1. Введение в Pyrhon для ML-инженеров и математиков.

Пайтон -- это высокоуровневый язык программирования с простым понятным синтаксисом, который стал стандартом де-факто в области машинного обучения, анализа данных и научных вычислений. Его философия делает акцент на читаемости кода, а богатая экосистема библиотек (таких как Numpy, Pandas, scikit-learn) делает его особенно мощным инструментом для математиков и ML-инженеров.
