```python
# task 6
address = "г. Бишкек, ул. Абдрахманова, д. 25"
parts = address.split(",")
city = ""
street = ""
house_number = ""

for p in parts:
    p = p.strip()
    
    if p.startswith("г."):
        city = p.replace("г. ", "")
        
    elif p.startswith("ул."):
        street = p.replace("ул. ", "")
        
    elif p.startswith("д."):
        house_number = int(p.replace("д. ", ""))
        
print("city: ", city)
print("street: ", street)
print("house_number: ", house_number)
```

```python
# task 19
boys = 20
girls = 25

best_sum = -1
best_split = None

for g1 in range(0, 23):
    b1 = 22 - g1
    g2 = girls - g1
    b2 = boys - b1
    
    if b1 < 0 or g2 < 0 or b2 < 0:
        continue
    if g2 + b2 != 23:
        continue
    
    p1 = g1 / 22 * 100
    p2 = g2 / 23 * 100
    s = p1 + p2
    
    if s > best_sum:
        best_sum = s
        best_split = (g1, b1, g2, b2)
        
print("max summa procentov: ", best_sum)
print("raspredelenie:", best_split)
```

```python
# task 21
v1 = 40
v2 = 30
d1_0 = 5
d2_0 = 3

min_dist = float('inf')
best_time_min = 0

for t_min in range(0, 60):
    t = t_min / 60
    x = d1_0 - v1 * t
    y = d2_0 - v2 * t
    dist = (x**2 + y**2) ** 0.5
    
    if dist < min_dist:
        min_dist = dist
        best_time_min = t_min
        
print("min rasstoyaniye: ", min_dist, "km")
print("dostigaetsya cherez: ", best_time_min, "minut")
```
