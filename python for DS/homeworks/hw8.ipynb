```python
# task 4
dict4 = {'a': 1, 'b': 4, 'c': 1, 'd': 5, 'e': 6}
dict4 = {k: v for k, v in dict4.items() if v >= 4}
print(dict4)
```

```python
# 5
dict5 = {1: {'a': 1}, 2: {'b': 4}, 3: {'c': 1}, 4: {'d': 5}, 5: {'e': 6}}
res = {outer_key: list(inner_dict.keys())[0] for outer_key, inner_dict in dict5.items()}
print(res)
```
