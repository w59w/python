```python
# task 6
def mult(nums):
    result = 1
    for n in nums:
        result *= n
    return result
print(mult([1, 2, 3, 4, 5]))
```

```python
# task 8
bd = ["Emil", "Kana", "Rena", "Aliya", "Nur"]
def check(name):
    if name in bd:
        return "yes"
    else:
        return "no"
print(check("Emil"))   # yes
print(check("Aibek"))  # no
```

```python
# task 10
def employees(users):
    result = []
    for user in users:
        if user["age"] > 21 and "IT" in user["work"]:
            result.append(user['name'])
        return result
users = [
    { 'name': 'Ron', 'age': 18, 'work': 'IT-backend developer'},
    { 'name': 'Sam', 'age': 23, 'work': 'Doctor'},
    { 'name': 'Bob', 'age': 26, 'work': 'Driver'},
    { 'name': 'John', 'age': 39, 'work': 'Teacher'},
    { 'name': 'Jack', 'age': 29, 'work': 'IT-frontend developer'}
]

print(employees(users))
```

```python
# task 12
def check_nan(lst):
    if "nan" in lst:
        lst = [x for x in lst if x != "nan"]
    return lst
a = [1, 2, 'nan', 3, 4]
print(check_nan(a))

```
